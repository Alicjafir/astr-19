```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def exp_function(x):
    return np.exp(x)
```

```python
y = exp_function(x)
```

```python
plt.figure(figsize=(8,5))
plt.plot(x,y)
plt.xlabel("Time [multiseconds]")
plt.ylabel("Awesomeness")
plt.savefig("MyFigure.pdf", format='pdf')
```
